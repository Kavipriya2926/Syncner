```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("D:/house_data.csv")
correlation = df['Price'].corr(df['lot area'])
spearman_corr, _ = stats.spearmanr(df['Price'], df['lot area'])
print(f"Correlation value:{correlation:.2f}")
sns.scatterplot(x='Price',y='lot area',data=df)
plt.title("Scatter plot of GrLivArea vs SalePrice")
plt.xlabel("Sale Price (Price)")
plt.ylabel("Above Ground Living Area (Area)")
plt.show()
```
