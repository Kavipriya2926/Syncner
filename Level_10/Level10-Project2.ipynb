```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
df=pd.read_csv("D:/E-commerce.csv")
print(df.head())
price_range = np.max(df['Price (Rs.)']) - np.min(df['Price (Rs.)'])
variance=np.var(df['Price (Rs.)'])
standard_deviation=np.std(df['Price (Rs.)'])
print(f"Range:{price_range}")
print(f"Variance:{variance}")
print(f"Standard deviation:{standard_deviation}")
sns.histplot(x='Price (Rs.)', data=df)
plt.title("Price Distribution of product")
plt.xlabel("Price")
plt.show()
sns.boxplot(x='Price (Rs.)', data=df)
plt.title("Price Distribution of product")
plt.xlabel("Price")
plt.show()
```
