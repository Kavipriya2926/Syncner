```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

data = pd.read_csv('D:/Mall_Customers.csv')  
data = data[['Age', 'Annual Income (k$)']]  
data = data.dropna()

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_scaled)

data['Cluster'] = kmeans.labels_
plt.scatter(data['Age'], data['Annual Income (k$)'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Clusters of Customers')
plt.show()

sil_score = silhouette_score(data_scaled, kmeans.labels_)
print(f'Silhouette Score: {sil_score}')

for n_clusters in range(2, 6):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_scaled)
    sil_score = silhouette_score(data_scaled, kmeans.labels_)
    print(f'Silhouette Score for {n_clusters} clusters: {sil_score}')
```
