```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("D:/online_retail.csv")
df = df.dropna(subset=['Quantity', 'UnitPrice'])
df['TotalSpent'] = df['Quantity'] * df['UnitPrice']
mean = np.mean(df['TotalSpent'])
median = np.median(df['TotalSpent'])
mode = df['TotalSpent'].mode()[0]  
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
df['SpendingCategory'] = np.where(df['TotalSpent'] < mean, 'Low Spender', 'High Spender')
print(df['SpendingCategory'].value_counts())
sns.countplot(x='SpendingCategory', data=df)
plt.title('Customer Segmentation: High vs Low Spenders')
plt.xlabel('Spending Category')
plt.ylabel('Number of Customers')
plt.show()
plt.figure(figsize=(8, 6))
sns.histplot(df['TotalSpent'], bins=30, kde=True)
plt.title('Distribution of Total Spending')
plt.xlabel('Total Spending')
plt.ylabel('Frequency')
plt.show()
df.to_csv('segmented_customers.csv', index=False)
```
