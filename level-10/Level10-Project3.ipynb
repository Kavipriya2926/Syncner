```python
import pandas as pd 
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("D:/online_retail.csv")
print(df)
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
uk_sales = df[df['Country'] == 'United Kingdom']['TotalSales']
france_sales = df[df['Country'] == 'France']['TotalSales']
t_stat, p_value = stats.ttest_ind(uk_sales, france_sales)
alpha = 0.05  
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between Region A and Region B sales.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between Region A and Region B sales.")
print(f"P-value: {p_value}")
sns.barplot(x='Country', y='TotalSales', data=df)
plt.xticks(rotation=90)
plt.show()
```
