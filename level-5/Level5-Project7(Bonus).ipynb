```python
import pandas as pd
from dataclasses import dataclass
import matplotlib.pyplot as plt
@dataclass
class DatasetStats:
    mean: float
    max: float
    min: float
    def plot_histogram(self, df, column_name):
        plt.figure(figsize=(8, 6))
        df[column_name].hist(bins=20, color='skyblue', edgecolor='black')
        plt.title(f"Histogram of {column_name}")
        plt.xlabel(column_name)
        plt.ylabel('Frequency')
        plt.show()
df = pd.read_csv("D:/house_data.csv")
column_name = 'Price' 
mean_value = df[column_name].mean()
max_value = df[column_name].max()
min_value = df[column_name].min()
stats = DatasetStats(mean=mean_value, max=max_value, min=min_value)
print(f"Mean: {stats.mean}, Max: {stats.max}, Min: {stats.min}")
def console_menu():
    print("Welcome to the Dataset Dashboard!")
    print("Choose an option:")
    print("1. Display stats")
    print("2. Plot histogram")
    print("3. Exit")
    while True:
        choice = input("Enter your choice (1/2/3): ")
        if choice == '1':
            print(f"Mean: {stats.mean}")
            print(f"Max: {stats.max}")
            print(f"Min: {stats.min}")
        elif choice == '2':
            column_name = input("Enter column name to plot (e.g., 'Price'): ")
            if column_name in df.columns:
                stats.plot_histogram(df, column_name)
            else:
                print("Invalid column name!")
        elif choice == '3':
            print("Exiting the dashboard.")
            break
        else:
            print("Invalid choice! Please try again.")
console_menu()
```
